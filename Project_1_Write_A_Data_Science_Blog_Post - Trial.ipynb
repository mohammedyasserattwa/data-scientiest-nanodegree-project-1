## Data Filtering by Country

In this section, we define a reusable function `filter_dataframe_by_country` that allows us to filter a dataframe based on a specific country or a list of countries. The function takes the following parameters:

- `df`: The dataframe to be filtered.
- `country_column`: The column in the dataframe containing country information.
- `target_country`: The target country or a list of countries to filter by.
- `columns_to_filter`: A list of columns to select after filtering.

The function then filters the dataframe based on the specified country or countries and selects the desired columns. It returns a filtered dataframe.

We demonstrate the function's usage by filtering the main dataframe `df` for data related to Egypt. Additionally, we exclude Egypt from the list of other countries before filtering.

This approach offers flexibility in filtering the data based on different countries and analyzing subsets of the dataset.


```python
def filter_dataframe_by_country(df, country_column, target_country, columns_to_filter):
    """
    Filter a dataframe by a specific country and select desired columns.
    
    Args:
    df (pd.DataFrame): The dataframe to filter.
    country_column (str): The column containing country information.
    target_country (str or list): The country or list of countries to filter by.
    columns_to_filter (list): List of columns to select after filtering.
    
    Returns:
    pd.DataFrame: The filtered and selected dataframe.
    """
    if isinstance(target_country, str):
        target_country = [target_country]
    
    filtered_df = df[df[country_column].isin(target_country)].dropna(subset=columns_to_filter)
    return filtered_df

# Filter the dataframe for Egypt and select specific columns.
columns_of_interest = ['LanguageHaveWorkedWith', 'LanguageWantToWorkWith']
egypt_df = filter_dataframe_by_country(df, 'Country', 'Egypt', columns_of_interest)
```

## Data Column Splitting

In this section, we introduce a function `split_column` that efficiently splits a specific column in a dataframe by the delimiter ';' and returns a split series. The function takes the following parameters:

- `df`: The dataframe containing the column to split.
- `column`: The column to be split.

The function then uses the Pandas built-in `str.split()` method to split the specified column by ';' and returns the resulting series.

We demonstrate the function's utility by splitting the language-related columns for Egypt. This approach allows us to work with more granular data and facilitates analysis based on individual language entries. The split data can be processed further for insights and visualizations.


```python
def split_column(df, column):
    """
    Split a column in a dataframe by ';' and return a split series.
    
    Args:
    df (pd.DataFrame): The dataframe to split.
    column (str): The column to split.
    
    Returns:
    pd.Series: The split series.
    """
    column_series = df[column].str.split(';')
    return column_series

# Splitting the dataframe for Egypt by specific columns into series.
egypt_have_worked_with = split_column(egypt_df, 'LanguageHaveWorkedWith')
egypt_want_to_work_with = split_column(egypt_df, 'LanguageWantToWorkWith')
```

## Flattening Nested Lists

In this section, we define a function `flatten_nested_list` that takes a nested list as input and returns a flat list by removing any nesting. The function takes the following parameter:

- `nested_list`: The nested list to be flattened.

The function uses a list comprehension to iterate through the nested list and strip each element before appending it to the flat list.

We demonstrate the function's effectiveness by applying it to the language-related lists for Egypt. By flattening the nested structure, we create lists of individual programming languages for analysis and visualization. This transformation allows us to work with the data more effectively and gain insights at a more granular level.


```python
def flatten_nested_list(nested_list):
    """
    Flatten a nested list, returning a flat list.
    
    Args:
    nested_list (list of lists): The nested list to be flattened.
    
    Returns:
    list: The flattened list.
    """
    flat_list = [obj.strip() for row in nested_list for obj in row]
    return flat_list

# Flattening the series into lists.
egypt_have_worked_with = flatten_nested_list(egypt_have_worked_with)
egypt_want_to_work_with = flatten_nested_list(egypt_want_to_work_with)
```

## Grouping and Creating Dictionaries

In this section, we define a function `group_list_to_dict` that groups a list of programming languages by their count and creates a list of dictionaries, each containing information about the programming language, its count, and the associated year. The function takes the following parameters:

- `data_list`: The list to be grouped and counted.
- `year`: The year associated with the data.

The function uses the `Counter` class to count occurrences of each programming language and then constructs a list of dictionaries with the desired information.

We demonstrate the function's utility by applying it to the flattened lists of programming languages for Egypt. By grouping the data and creating dictionaries, we create structured data suitable for analysis and visualization. This transformation allows us to see the popularity of programming languages in different regions for a specific year.


```python
def group_list_to_dict(data_list, year):
    """
    Group a list by count and create a list of dictionaries with language, count, and year.
    
    Args:
    data_list (list): The list to be grouped and counted.
    year (int): The year associated with the data.
    
    Returns:
    list of dict: List of dictionaries with language, count, and year.
    """
    grouped_list = dict(Counter(data_list))
    grouped_data = [{'Programming Language': key, 'Count': value, 'Year': year} for key, value in grouped_list.items()]
    return grouped_data

# Grouping the lists into dictionaries for Egypt.
egypt_have_worked_with_grouped = group_list_to_dict(egypt_have_worked_with, 2023)
egypt_want_to_work_with_grouped = group_list_to_dict(egypt_want_to_work_with, 2023)
```

## Calculating Percentage of Counts

In this section, we define a function `calculate_percentage` that calculates the percentage of counts within a specified column of a dataframe. The function takes the following parameters:

- `df`: The dataframe containing the data.
- `column`: The column for which to calculate the percentage.

The function performs vectorized calculations to calculate the percentage of each count within the column, scaled based on the total count.

We then apply this function to calculate the percentage of counts for each programming language within the specified columns for Egypt. The calculated percentages are stored in new columns named 'Percentage' within the respective dataframes.

By calculating percentages, we gain insight into the relative popularity of different programming languages. This information is valuable for making comparisons and identifying trends in language preferences.


```python
def calculate_percentage(df, column):
    """
    Calculate the percentage of counts within a column and return scaled data.
    
    Args:
    df (pd.DataFrame): The dataframe containing the data.
    column (str): The column for which to calculate the percentage.
    
    Returns:
    pd.Series: The scaled data representing percentages.
    """
    column_data = df[column]
    percentage_data = column_data / column_data.sum()
    return percentage_data

# Calculating the percentage of counts for each year.
egypt_worked['Percentage'] = calculate_percentage(egypt_worked, 'Count')
egypt_wanted['Percentage'] = calculate_percentage(egypt_wanted, 'Count')
```

## Question 3: How is the utilization of programming languages at work connected to the programming languages that individuals aspire to work with in Egypt, as revealed by the Stack Overflow survey data for the year 2023?

```python
# Create a list to store dictionaries representing programming language transitions
rows = []

# Iterate through pairs of worked with and wanted to work with languages
for pair in zip(egypt_have_worked_with, egypt_want_to_work_with):
    for lang in pair:
        # Append dictionaries to the list with worked and wanted languages
        rows.append({
            'worked_programming_languages': pair[0], 
            'wanted_programming_languages': Counter(pair)
        })
        
# Create a dataframe from the list of dictionaries and group by worked languages
programming_language_transition = pd.DataFrame(rows).groupby('worked_programming_languages').agg({'wanted_programming_languages': 'sum'}).reset_index()
```

## Creating a Transition Matrix for Programming Languages

In this section, we create a transition matrix that represents the relationship between worked programming languages and wanted programming languages. Each cell in the matrix indicates the proportion of developers who have worked with a specific language and also want to work with another language.

To achieve this, we iterate through the rows of the `programming_language_transition` dataframe. For each row, we calculate the total count of wanted programming languages and then update the corresponding cells in the matrix with the calculated proportions.

The resulting matrix provides insights into the patterns of transitions between programming languages based on developers' work and desire to work with certain languages.

Let's display the first few rows of this transition matrix to visualize the captured relationships.


```python
# Create a matrix to represent language transitions
for row in programming_language_transition['worked_programming_languages']:
    programming_language_transition[row] = 0

for index, row in programming_language_transition.iterrows():
    total_wanted = sum(row['wanted_programming_languages'].values())
    for key, value in row['wanted_programming_languages'].items():
        try:
            programming_language_transition.at[index, key] = value / total_wanted
        except:
            continue
```

## Visualizing Programming Language Transitions with Heatmap

In this section, we create a heatmap to visualize the transition matrix that represents the relationship between worked programming languages and wanted programming languages. Each cell in the heatmap indicates the proportion of developers who have worked with a specific language and also want to work with another language.

We create the transition matrix by excluding the 'wanted_programming_languages' column and setting the 'worked_programming_languages' as the index. This matrix captures the proportions of transitions between programming languages.

We then create a heatmap using the Seaborn library, color-coded with the "YlGnBu" color map. Each cell is annotated with the calculated proportion values, enhancing readability. The heatmap provides an intuitive way to comprehend the language transition patterns.

Let's display the heatmap to visualize the relationships between programming languages in terms of transitions.


```python
# Create a matrix to represent language transitions
matrix = programming_language_transition.drop('wanted_programming_languages', axis=1)\
                                        .set_index('worked_programming_languages')

# Create a heatmap to visualize the transition matrix with modified colors
plt.figure(figsize=(16, 10))
sns.heatmap(matrix, cmap="RdYlGn", center=0, linewidths=0.5)

# Set title and labels
plt.title('Programming Languages Transition Heatmap', size=20)
plt.xlabel('Wanted Programming Languages', fontsize=14)
plt.ylabel('Worked Programming Languages', fontsize=14)

plt.show()
```
